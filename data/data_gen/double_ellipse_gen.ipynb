```python
import pandas as pd
import random

# Set the random seed for reproducibility
random.seed(48)

# Number of data points
num_data_points = 5000000


# Number of segments for each variable
num_segments = 10

# Define the ranges for each parameter
A_range = (240, 450)
B_range = (60, 200)
theta_range = (0, 45)
phi_range = (0, 360)
px_range = (300, 800)
py_range = (300, 800)

# Generate random data
data = {
    'A': [],
    'B': [],
    'theta': [],
    'Phi': [],
    'Px': [],
    'Py': [],
}


# Generate remaining random data
for _ in range(num_data_points):
    data['A'].append(random.randint(*A_range))
    data['B'].append(random.randint(*B_range))
    data['theta'].append(random.randint(*theta_range))
    data['Phi'].append(random.randint(*phi_range))
    data['Px'].append(random.randint(*px_range))
    data['Py'].append(random.randint(*py_range))

# Create a DataFrame
df = pd.DataFrame(data)
```

```python
import pandas as pd
import numpy as np


def filter_data(df):
    """
    过滤数据的函数。

    参数：
    df (DataFrame): 包含原始数据的DataFrame。

    返回：
    DataFrame: 过滤后的DataFrame。
    """
    # 定义过滤条件
    def condition1(df):
        return df['A'] < df['Py']

    def condition2(df):
        return df['B'] < (df['Px'] / 2)
    

    # 使用逻辑与连接所有过滤条件
    combined_condition = pd.Series(True, index=df.index)
    for condition in [condition1,condition2]:
        combined_condition = combined_condition & condition(df)

    # 根据过滤条件进行筛选
    filtered_df = df[combined_condition]

    return filtered_df

# 读取数据
# df = pd.read_csv('random_data_with_segments.csv')

# 使用函数过滤数据
filtered_df = filter_data(df)

# 打印过滤前后的数据数量
print(f"过滤前的数据数量: {len(df)}")
print(f"过滤后的数据数量: {len(filtered_df)}")

# 如果需要保存过滤后的数据，可以使用以下代码
# filtered_df.to_csv('filtered_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import os
import pandas as pd
import numpy as np
import matplotlib.transforms as transforms
# import cv2
import math
from tqdm import tqdm
```

```python
def double_ellipse_draw(save_name, A, B, a, Px, Py,phi, dir):
    fig, ax = plt.subplots(figsize=(1, 1))
    ax.set_xlim(-450, 450)
    ax.set_ylim(-450, 450)

    # 绘制白色背景
    ax.set_facecolor('white')

    # 绘制黄色矩形
    rect_center = (0, 0)  # 矩形中心点坐标
    rect_width = Px  # 矩形长度
    rect_height = Py  # 矩形宽度
    rect = patches.Rectangle((rect_center[0] - rect_width / 2, rect_center[1] - rect_height / 2),
                             rect_width, rect_height, linewidth=1, edgecolor='yellow', facecolor='yellow')
    ax.add_patch(rect)

    # 绘制红色矩形1
    ellipse_center = (Px/4, 0)  # 椭圆中心点坐标
    major_axis = A  # 长轴
    minor_axis = B  # 短轴
    angle = -a  # 旋转角度
    ellipse_1 = patches.Ellipse(ellipse_center, major_axis, minor_axis, angle=angle, linewidth=1, edgecolor='red', facecolor='red')

    # 绘制红色矩形2
    ellipse_center = (-Px/4, 0)  # 椭圆中心点坐标
    major_axis = A  # 长轴
    minor_axis = B  # 短轴
    angle = a  # 旋转角度
    ellipse_2 = patches.Ellipse(ellipse_center, major_axis, minor_axis, angle=angle, linewidth=1, edgecolor='red', facecolor='red')


    # 将两个红色矩形作为一个整体，围绕画布中心进行旋转
    trans = patches.Affine2D().rotate_deg(phi) + ax.transData
    ellipse_1.set_transform(trans)
    ellipse_2.set_transform(trans)

    ax.add_patch(ellipse_1)
    ax.add_patch(ellipse_2)

    # 显示图像
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis('off')  # 关闭坐标轴
    plt.subplots_adjust(top=1, bottom=0, right=1, left=0, hspace=0, wspace=0)
    plt.margins(0, 0)
    plt.savefig(os.path.join(dir, save_name + '.png'))
    # plt.show()
    plt.clf ()
    plt.close()

# double_ellipse_draw("1",120,300,17.5,500,350,30,os.getcwd())
```

```python
save_dir = r'D:\simulation\BMD\double_ellipse\pic_expand'
for i in tqdm(range(len(data))):
    num = 0
    name = os.path.splitext(data.iloc[i]['Pic'])[0]
    if os.path.exists(os.path.join(save_dir, name + '.png')):
        continue
    else:
        if num == 100000:
            print('done')
            break
        a = data.iloc[i]['A']
        b = data.iloc[i]['B']
        Px = data.iloc[i]['Px']
        Py = data.iloc[i]['Py']
        theta = data.iloc[i]['theta']
        phi = data.iloc[i]['Phi']
        double_ellipse_draw(name,int(a), int(b), int(theta), int(Px), int(Py), int(phi), save_dir)
        plt.close("all")
        num += 1
```

```python
from PIL import Image
# Image.open(os.path.join(droplist["dir"][1520],droplist["name"][1520]+".png"))

import cv2
import numpy as np

def check_check(img,Px,Py):

    # 读取图像
    image = cv2.imread(img)

    # 计算图像中心坐标
    image_center_x = image.shape[1] // 2
    image_center_y = image.shape[0] // 2


    # 定义矩形框尺寸
    rect_width, rect_height = Px//9, Py//9

    # 计算矩形框左上角坐标
    rect_x = image_center_x - rect_width // 2
    rect_y = image_center_y - rect_height // 2

    # 定义红色
    red = [0, 0, 255]

    # 检查框外的红色像素
    is_red_outside = False
    for y in range(image.shape[0]):
        for x in range(image.shape[1]):
            # 检查是否在框外

            if not (rect_x <= x < rect_x + rect_width and rect_y <= y < rect_y + rect_height):
                # 检查像素是否为红色
                if np.array_equal(image[x,y], red):
                    is_red_outside = True
                    break
        if is_red_outside:
            break

    # 输出结果
    if is_red_outside:
        return True
    else:
        return False

    # 绘制图像和矩形框（可选）
    # cv2.rectangle(image, (rect_x, rect_y), (rect_x + rect_width, rect_y + rect_height), (0, 255, 0), 2)
    # cv2.imshow('Image with Rectangle', image)
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()
```
